Q1. For a task with three states $s_1,s_2,s_3$ where $s_i \in \mathbb{R}$ generate the feature vectors using

(a) order-3 polynomial-bases $x_i(s) = \prod_{j=1}^k s_j^{c_{i,j}}$ 
> Answer:

```python
import numpy as np
#this function create c matrix
def create_c(k,n):
    c=[[] for j in range(k)]
    for i in range(k):
        for j in range(int(np.power(n+1,i))):
            for p in range(n+1):
                for r in range(int(np.power(n+1,k-i-1))):
                    c[i].append(p)
    c=np.array(c,dtype=np.int8)
    return c


def sup(i): #create superscript
    if i == 1:
        a=chr(0x00b9)
    elif 2 <= i <= 3:
        a=chr(0x00b0 + i)
    else:
        a=chr(0x2070 + i)
    return a

def sub(i): #create subscript
    a=chr(0x2080 + i)
    return a

def polinomial_bases(k,n):
    c=create_c(k,n)
    d=np.power((n+1),k)
    state=[]
    for i in range(k):
        state.append("S"+sub(i+1))

    x_i=[]
    for j in range(d):
        sum_=""
        for i in range(k):
            if c[i][j]==0:
                pass
            elif c[i][j]==1:
                sum_+=state[i]
            else:
                sum_+=state[i]+sup(c[i][j])
        if len(sum_)==0:
            sum_="1"
        x_i.append(sum_)
    print("length x_i(s) = "+ str(len(x_i)))
    print(x_i)
```

```python
polinomial_bases(3,3)
```

(b) order-3 Fourier-bases $x_i(s)= \cos(\pi \mathbf{s}^{\top}\mathbf{C}^i)$

```python
def Fourier_bases(k,n):
    d=np.power((n+1),k)
    c=create_c(k,n)
    state=[]
    for i in range(k):
        state.append("S"+sub(i+1))
    x_i_f=[]
    for j in range(d):
        sum_=""
        for i in range(k):
            if c[i][j]==0:
                pass
            elif c[i][j]==1:
                if len(sum_)==0:
                    sum_+=state[i]
                else:
                    sum_+="+"+state[i]
            else:
                if len(sum_)==0:
                    sum_+=str(c[i][j])+state[i]
                else:
                    sum_+="+"+str(c[i][j])+state[i]
        if len(sum_)==0:
            x_i_f.append("1")
        else:
            x_i_f.append("cos(\u03C0("+sum_+"))")
    print("length x_i(s) = "+ str(len(x_i_f)))
    print(x_i_f)
```
