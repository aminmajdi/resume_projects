```python
# Write a function that plays a given k-armed bandit problem
# this function should include greedy, epsilon-greedy and UCB action selection strategies.
# You can include all strategies in one function or write two functions (one for epsilon-greedy and another for UCB)
# your implementation should return the selected action and the reward gained by selecting that action

# Your code here
def play_k_armed_bandit(e,c):
    Q_star=k_armed_bandit()
    k=7
    optimal_value_index=np.empty(shape=(500,1),dtype='object')
    A = np.empty(shape=(500,1000),dtype='object')
    revard = np.empty(shape=(500,1000),dtype='object')
    is_optimal=np.empty(shape=(500,1000),dtype='object')
    for i in range(500):
        q=Q_star[i][:]
        optimal_value_index[i]=np.argmax(q)
        Q_t=[0]*k
        action_counter=[0]*k # how many time the action is selected
        for j in range(1000):
            if c == 0:
                if np.random.rand(1)> e:
                    action_index=np.argmax(Q_t)
                else:
                    action_index=Q_t.index(np.random.choice(Q_t))
            elif c!=0:
                if j<k:
                    for x in range(k):
                        if action_counter[x]==0:
                            action_index=x
                            break
                        else: 
                            continue
                else:
                    action_index=np.argmax(Q_t+c*np.sqrt(np.log(j)/action_counter[np.argmax(Q_t)]))
                
            A[i][j]=action_index
            action_counter[action_index]+=1 
            revard[i][j]=np.random.normal(q[action_index],1)
            Q_t[action_index]=Q_t[action_index]+(1/action_counter[action_index])*(revard[i][j]-Q_t[action_index])
            if action_index ==optimal_value_index[i]:
                is_optimal[i][j]=1
            else:
                is_optimal[i][j]=0
    return(revard,is_optimal)  
    
```

```python
# Average the rewards and number of best action selections over all runs
# then plot (a) average reward, (b) %optimal action over all time steps
# you should plot two subplots (a and b) each showing all strategies

# Your code here
def average(reward,is_optimal):
    average_reward=[0]*1000
    optimal_average=[0]*1000
    for i in range(1000):
        average_reward[i]=sum(reward[:,i])/len(reward[:,i])
        optimal_average[i]=(np.sum(is_optimal[:,i])/len(is_optimal[:,i]))*100
        
    return(average_reward,optimal_average)


reward,optimal=play_k_armed_bandit(0,0)
reward_greedy,optimal_average_greedy=average(reward,optimal)

reward,optimal=play_k_armed_bandit(0.1,0)
reward_egreedy_1,optimal_average_egreedy_1=average(reward,optimal)

reward,optimal=play_k_armed_bandit(0.01,0)
reward_egreedy_01,optimal_average_egreedy_01=average(reward,optimal)

reward,optimal=play_k_armed_bandit(0,1)
reward_UCB_1,optimal_average_UCB_1=average(reward,optimal)

reward,optimal=play_k_armed_bandit(0,2)
reward_UCB_2,optimal_average_UCB_2=average(reward,optimal)

fig=plt.figure()

plt.subplot(1,2,1)
plt.xlabel("time_step")
plt.ylabel("average_reward")
plt.plot(range(1000),reward_greedy,label='greedy')
plt.plot(range(1000),reward_egreedy_1,label='e=0.1')
plt.plot(range(1000),reward_egreedy_01,label='e=0.01')
plt.plot(range(1000),reward_UCB_1,label='UCB_1')
plt.plot(range(1000),reward_UCB_2,label='UCB_2')

plt.legend()

plt.subplot(1,2,2)
plt.xlabel("time_step")
plt.ylabel("optimal_average")
plt.plot(range(1000),optimal_average_greedy,label='greedy')
plt.plot(range(1000),optimal_average_egreedy_1,label='e=0.1')
plt.plot(range(1000),optimal_average_egreedy_01,label='e=0.01')
plt.plot(range(1000),optimal_average_UCB_1,label='UCB_1')
plt.plot(range(1000),optimal_average_UCB_2,label='UCB_2')

plt.legend()
plt.tight_layout()

plt.show()
```
