```python
import numpy as np
import matplotlib.pyplot as plt
# Your code here
def play(p_h,teta):
    
    sweep=0
    v=[0]*101
    v[100]=1
    final_policy=[0]*100
    delta=1
    empty=np.empty(shape=(1,101))
    i_p=[0]*100
    while delta>teta:
        if sweep==0:
            value=v
            value_mat=v
        else:
            value=value_mat[sweep,:]
        sweep+=1
        delta = 0
        for state in range(1,100):
            v_reserve=value[state]
            v_action=0
            v_array=[]
            for action in range(1,min(state+1,100-state+1)):
                v_action=p_h*value[state+action]+(1-p_h)*value[state-action]
                v_array.append(v_action)
            value[state]=max(v_array)
#            print(max(v_array))
            if i_p[state]!=1:
                if final_policy[state]!=np.argmax(v_array)+1:
                    final_policy[state]=np.argmax(v_array)+1
                else:
                    i_p[state]+=1
            delta=max(delta,abs(v_reserve-value[state]))

        value_mat=np.vstack((value_mat,value))
#                if v_action_reserve<v_action:
#                    w[sweep,state]=v_action
#                    final_policy[state]=action
#                    print(w[sweep,state])
            
#            print(w[sweep,state])
#            delta=max(delta,abs(v_reserve-w[sweep,state]))
#            print(delta)

    return(value_mat,final_policy,sweep)

# Your code here

value,final_policy,s=play(0.4,0.00000000004)
#print(final_policy)
print(final_policy)


fig=plt.figure()
plt.subplot(1,2,1)
plt.xlabel("capital")
plt.ylabel("value_estimates")
for counter in range(s+1):
    plt.plot(range(101),value[counter,:],label='sweep'+str(counter+1))
plt.legend()

plt.subplot(1,2,2)
plt.xlabel("capital")
plt.ylabel("final policy")
plt.step(range(100),final_policy,label='action')

plt.legend()

    
```
