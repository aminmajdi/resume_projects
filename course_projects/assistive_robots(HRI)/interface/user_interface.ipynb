```python
from tkinter import *
from PIL import ImageTk,Image

root = Tk()
root.title('Assistive Robot')


frame3 = LabelFrame(root, padx=50, pady=20)
frame3.pack(padx=10, pady=10)

my_img = ImageTk.PhotoImage(Image.open("amin.png"))
my_label = Label(frame3,image=my_img)
my_label.grid(row=0)

frame4 = LabelFrame(root, padx=50, pady=20)
frame4.pack(padx=10, pady=10)    
label2 = Label(frame4 , text = "What should I do?",font=("Arial", 40))#.pack()
label2.pack()

frame5 = LabelFrame(root, padx=50, pady=20)
frame5.pack(padx=10, pady=10)
def open_the_door():
    label_open = Label(frame5 , text = 'The door is open and Amin is comming in',fg='blue',font=("Arial", 40))
    label_open.grid(row=1)
    
def go_to_home():
    label_open = Label(frame5 , text = 'ok I am heading to my home position',fg='blue',font=("Arial", 40))
    label_open.grid(row=1)
    
frame2 = LabelFrame(root, padx=50, pady=20)
frame2.pack(padx=10, pady=10)

button_open = Button(frame2, text="Open the door",font=("Arial", 40),padx = 20,pady = 20 ,fg = "black",bg='green', command=open_the_door)

button_home = Button(frame2, text="Do nothing",font=("Arial", 40),padx = 20,pady = 20 ,fg = "black",bg='green', command=go_to_home)

button_open.grid(row=0, column=0,padx=10, pady=10)
button_home.grid(row=0, column=1,padx=10, pady=10)


root.mainloop()
```
