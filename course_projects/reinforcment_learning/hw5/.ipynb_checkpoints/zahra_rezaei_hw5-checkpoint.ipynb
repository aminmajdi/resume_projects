The following Windy Gridworld includes a crosswind running upward through the middle of the grid. The strength of wind is shown under each column that you should use for shifting. ## Part I
Implement SARSA (on-policy TD control) using $\varepsilon$–greedy policy with parameters $\varepsilon=0.1, \alpha=0.5$, and $Q_0 (s,a)=0$ for all $s,a$. Apply your implementation to the windy gridworld in the following scenarios:


**(a)** In this scenario, the agent’s action set includes four moves (i.e. Up, Down, Right, Left). Plot the number of episodes vs. time steps (similar to graph in page 130). Then calculate the average episode length. 

**(b)** In this scenario, the agent has access to King’s moves (eight actions). You need to define a new action set and re-run the experiment in previous scenario. Plot the number of episodes vs. time steps. Then calculate the average episode length.

**(c)** In this scenario, you need to add a ninth action that causes no movement at all other than that caused by the wind. Plot the number of episodes vs. time steps. Then calculate the average episode length.

**(d)** In this scenario, the agent has access to King’s moves (eight actions). Assume that the effect of the wind, if there is any, is stochastic, sometimes varying by 1 from the mean values given for each column. That is, a third of the time you move exactly according to these values, as in the previous part, but also a third of the time you move one cell above that, and another third of the time you move one cell below that. For example, if you are one cell above the goal, one-third of the time you move two cells above the goal, and one-third of the time you move to the goal. Plot the number of episodes vs. time steps. Then calculate the average episode length.

**Note:** all four plots should be shown in one figure for comparison.



```python

def TD_sarsa(num_act,shift=False):
        wind=[0,0,0,1,1,1,2,2,1,0]
        num_play=170
        num_col=10
        num_row=7
        alpha=0.5
        epsilon=0.1
        reward =-1
        action_value=[]
        time_step=[]
        step_average=[]
        for x in range(num_row):
            action_value.append([])
            for y in range(num_col):
                action_value[x].append([])
                for z in range(num_act):
                    action_value[x][y].append(0)
        
        for x in range(num_play):
            time_step.append(0)
            step_average.append(0)
            
        
        action_dictionary={
            0:[0,1],
            1:[1,0],
            2:[0,-1],
            3:[-1,0],
            4:[1,1],
            5:[1,-1],
            6:[-1,-1],
            7:[-1,1],
            8:[0,0]
        }
        episode=0
        while episode<num_play:
            episode+=1
            state=[3,0]
            goal=[3,7]
            s_row=state[0]
            s_col=state[1]
            
            if np.random.rand(1)> epsilon:
                action_index=np.argmax(action_value[s_row][s_col])
            else:
                action_index=np.random.choice(range(num_act))
                
            action=action_dictionary[action_index]
            
            step=0
            not_done=True
            while not_done:
                step+=1
                if shift==False:
                    w_shift=wind[state[1]]
                else:
                    if wind[state[1]]==0:
                        w_shift=wind[state[1]]
                    else:
                        w_shift=wind[state[1]]+np.random.choice([-1,0,1])

                new_s_col=state[1]+action[1]
                if new_s_col>9:
                    new_s_col=9
                if new_s_col<0:
                    new_s_col=0
                
                new_s_row=state[0]+action[0]+w_shift
                if new_s_row>6:
                    new_s_row=6
                if new_s_row<0:
                    new_s_row=0
                
                new_state= [new_s_row,new_s_col]
                if np.random.rand(1)> epsilon:
                    new_action_index=np.argmax(action_value[new_s_row][new_s_col])
                else:
                    new_action_index=np.random.choice(range(num_act))
                new_action=action_dictionary[new_action_index]
                action_value[s_row][s_col][action_index]=action_value[s_row][s_col][action_index]+(alpha*(reward+action_value[new_s_row][new_s_col][new_action_index]-action_value[s_row][s_col][action_index]))

                action_index=new_action_index
                action=new_action                
                
                s_row=new_s_row
                s_col=new_s_col
                state=new_state

                if state ==goal:
                    time_step[episode-1]=step
                    not_done=False
        step_sum=[sum(time_step[0:i]) for i in range(len(time_step)+1)]
        c_step_sum=step_sum[1:]
        for i in range(num_play):
            step_average[i]=c_step_sum[i]/(i+1)

        return step_sum , step_average
    
step_sum1 , av1=TD_sarsa(4)
step_sum2 , av2=TD_sarsa(8)
step_sum3 , av3=TD_sarsa(9)
step_sum4 , av4=TD_sarsa(9,True)




fig = plt.figure(figsize=(10,20))

ax1 = fig.add_subplot(211)
num_play=170
ax1.plot(step_sum1,range(num_play+1),label='4')
ax1.plot(step_sum2,range(num_play+1),label='8')
ax1.plot(step_sum3,range(num_play+1),label='9')
ax1.plot(step_sum4,range(num_play+1),label='9 drifty')


ax1.set_xlabel('time_step')
ax1.set_ylabel('episode')
ax1.legend(fontsize=20)

ax2 = fig.add_subplot(212)

ax2.plot(av1,label='4')
ax2.plot(av2,label='8')
ax2.plot(av3,label='9')
ax2.plot(av4,label='9 drifty')


ax2.set_xlabel('episode')
ax2.set_ylabel('average episode length')
ax2.legend(fontsize=20)
```

## Part II
Consider the following gridworld with four actions (up, down, right, and left). If the action takes the agent off the gird, the agent stays in the same state. In each non-terminating step, the agent receives a random reward of -12 or +10 with equal probability. The reward for reaching the goal state is +5 and the episode ends when the agent reaches the goal.

Use $\varepsilon$–greedy policy with $\varepsilon(s) = \frac{1}{\sqrt{n(s)}}$ where $n(s)$ is the number of times state $s$ has been visited, assuring infinite exploration in the limit which is a theoretical requirement for the convergence of both **Q-learning** and **Double Q-learning**. 


a. Implement **Q-learning** and **Double Q-learning** and apply them to this problem for 10,000 experiments using the learning rate $\alpha=\frac{1}{n(s,a)}$.

b. Plot the average reward per step vs. number of time steps averaged over 10,000 experiments. The length of an episode following the optimal policy is five actions, so the optimal average reward per step is +0.2. Plot this true value in your figure and see how close your algorithm gets to the true value.

c. Plot the maximal action value in the starting state $S$ (i.e. $max_a Q(s,a)$) averaged over 10,000 experiments. The optimal value of maximally valued action in the starting state is $5\gamma^4 - \sum_{k=0}^3 \approx 0.36$. Plot this true value in your figure and see how close your algorithm gets to the true value.

d. Repeat the experiments with $\alpha = \frac{1}{n(s,a)^{0.8}}$ and redo steps b and c.


**Note:** You should have four figures (average rewards and maximal action values for different learning rates, $2 \times 2$). 


```python

def Q_learning(p):
        num_play=10000
        num_col=3
        num_row=3
        gamma=0.95
        action_value=[]
        n_s=[]
        n_sa=[]
        time_step=[]
        rewards_sum_list=[]
        acc_time_steps_list=[]
        acc_time_steps=0
        average_r_s=[]
        Q_max=[]
        Q_max_sum=[]
        
        for x in range(num_row):
            action_value.append([])
            n_sa.append([])
            for y in range(num_col):
                action_value[x].append([])
                n_sa[x].append([])
                for z in range(4):
                    action_value[x][y].append(0)
                    n_sa[x][y].append(0)
        
        for x in range(num_row):
            n_s.append([])
            for y in range(num_col):
                n_s[x].append(0)
        
        for x in range(num_play):
            time_step.append(0)
            acc_time_steps_list.append(0)
            average_r_s.append(0)
            Q_max_sum.append(0)
            

        action_dictionary={
            0:[0,1],
            1:[1,0],
            2:[0,-1],
            3:[-1,0],
        }
        episode=0
        while episode<num_play:
            episode+=1
            reward_list=[]
            state=[0,0]
            s_row=state[0]
            s_col=state[1]
            goal=[2,2]
            step=0
            n_done=True
            while n_done:
                step+=1
                n_s[s_row][s_col]+=1
                epsilon=1/np.sqrt(n_s[s_row][s_col])
                if np.random.rand(1)> epsilon:
                    action_index=np.argmax(action_value[s_row][s_col])
                else:
                    action_index=np.random.choice(range(4))
                action=action_dictionary[action_index]
                new_s_col=state[1]+action[1]
                if new_s_col>2:
                    new_s_col=2
                if new_s_col<0:
                    new_s_col=0
                
                new_s_row=state[0]+action[0]
                if new_s_row>2:
                    new_s_row=2
                if new_s_row<0:
                    new_s_row=0                
                new_state= [new_s_row,new_s_col]                
                
                if new_state==goal:
                    reward =5
                else:
                    if np.random.rand(1)>0.5:
                        reward =-12
                    else:
                        reward =10
                n_sa[s_row][s_col][action_index]+=1
                alpha=1/np.power(n_sa[s_row][s_col][action_index],p)
                new_action_index=np.argmax(action_value[new_s_row][new_s_col])
                action_value[s_row][s_col][action_index]+=(alpha*(reward+gamma*action_value[new_s_row][new_s_col][new_action_index]-action_value[s_row][s_col][action_index]))
                state=new_state

                s_row=new_s_row
                s_col=new_s_col                
                reward_list.append(reward)
                if step==100:
                    state =goal
                
                
                if state ==goal:
                    
                    Q_max.append(max(action_value[0][0]))
                    Q_max_sum[episode-1]=(np.sum(Q_max))/episode
                    
                    acc_time_steps+=step
                    acc_time_steps_list[episode-1]=acc_time_steps
                    rewards_sum_list.append(np.sum(reward_list))
                    average_r_s[episode-1]=np.sum(rewards_sum_list)/acc_time_steps
                    n_done=False
        return acc_time_steps_list , average_r_s , Q_max_sum                  

```

```python

def double_Q_learning(p):
        num_play=10000
        num_col=3
        num_row=3
        gamma=0.95
        action_value1=[]
        action_value2=[]
        n_s=[]
        n_sa1=[]
        n_sa2=[]
        time_step=[]
        rewards_sum_list=[]
        acc_time_steps_list=[]
        acc_time_steps=0
        average_r_s=[]
        Q_max=[]
        Q_max_sum=[]
        
        for x in range(num_row):
            action_value1.append([])
            n_sa1.append([])
            action_value2.append([])
            n_sa2.append([])
            for y in range(num_col):
                action_value1[x].append([])
                n_sa1[x].append([])
                action_value2[x].append([])
                n_sa2[x].append([])
                for z in range(4):
                    action_value1[x][y].append(0)
                    n_sa1[x][y].append(0)
                    action_value2[x][y].append(0)
                    n_sa2[x][y].append(0)
        
        for x in range(num_row):
            n_s.append([])
            for y in range(num_col):
                n_s[x].append(0)
        
        for x in range(num_play):
            time_step.append(0)
            acc_time_steps_list.append(0)
            average_r_s.append(0)
            Q_max_sum.append(0)
            

        action_dictionary={
            0:[0,1],
            1:[1,0],
            2:[0,-1],
            3:[-1,0],
        }
        episode=0
        while episode<num_play:
            episode+=1
            reward_list=[]
            state=[0,0]
            s_row=state[0]
            s_col=state[1]
            goal=[2,2]
            step=0
            n_done=True
            while n_done:
                step+=1
                n_s[s_row][s_col]+=1
                epsilon=1/np.sqrt(n_s[s_row][s_col])
                Q_av=[0,0,0,0]
                for i in range(4):
                    Q_av[i]=(action_value1[s_row][s_col][i]+action_value2[s_row][s_col][i])/2
                if np.random.rand(1)> epsilon:
                    action_index=np.argmax(Q_av)
                else:
                    action_index=np.random.choice(range(4))
                action=action_dictionary[action_index]
                new_s_col=state[1]+action[1]
                if new_s_col>2:
                    new_s_col=2
                if new_s_col<0:
                    new_s_col=0
                
                new_s_row=state[0]+action[0]
                if new_s_row>2:
                    new_s_row=2
                if new_s_row<0:
                    new_s_row=0                
                new_state= [new_s_row,new_s_col]                
                
                if new_state==goal:
                    reward =5
                else:
                    if np.random.rand(1)>0.5:
                        reward =-12
                    else:
                        reward =10
                if np.random.rand(1)>0.5:
                    n_sa1[s_row][s_col][action_index]+=1
                    alpha=1/np.power(n_sa1[s_row][s_col][action_index],p)
                    new_action_index=np.argmax(action_value1[new_s_row][new_s_col])
                    action_value1[s_row][s_col][action_index]+=(alpha*(reward+gamma*action_value2[new_s_row][new_s_col][new_action_index]-action_value1[s_row][s_col][action_index]))
                else:
                    n_sa2[s_row][s_col][action_index]+=1
                    alpha=1/np.power(n_sa2[s_row][s_col][action_index],p)
                    new_action_index=np.argmax(action_value2[new_s_row][new_s_col])
                    action_value2[s_row][s_col][action_index]+=(alpha*(reward+gamma*action_value1[new_s_row][new_s_col][new_action_index]-action_value2[s_row][s_col][action_index]))
                state=new_state
                

                s_row=new_s_row
                s_col=new_s_col                
                reward_list.append(reward)

                
                
                if state ==goal:
                    
                    Q_max.append(max([max(action_value1[0][0]),max(action_value2[0][0])]))
                    Q_max_sum[episode-1]=(np.sum(Q_max))/episode
                    
                    acc_time_steps+=step
                    acc_time_steps_list[episode-1]=acc_time_steps
                    rewards_sum_list.append(np.sum(reward_list))
                    average_r_s[episode-1]=np.sum(rewards_sum_list)/acc_time_steps
                    n_done=False
        return acc_time_steps_list , average_r_s , Q_max_sum                  
```

```python
NQ_1,avQ_1,maxQ_1=Q_learning(1)
N2Q_1,av2Q_1,max2Q_1=double_Q_learning(1)
NQ_08,avQ_08,maxQ_08=Q_learning(0.8)
N2Q_08,av2Q_08,max2Q_08=double_Q_learning(0.8)

num_play=10000
fig=plt.figure(figsize=(20,20))
ax1 = fig.add_subplot(221)

ax1.plot(NQ_1,avQ_1,label='Q and 1')
ax1.plot(N2Q_1,av2Q_1,label='double_Q and 1')
optimal_value=[0.5 for i in range(NQ_1[-1])]
ax1.plot(optimal_value,label='optimal_value')
ax1.set_xlabel('number of time steps',fontsize=20)
ax1.set_ylabel('average rewards',fontsize=20)
ax1.legend(fontsize=20)

ax2 = fig.add_subplot(222)

ax2.plot(NQ_08,avQ_08,label='Q and 0.8')
ax2.plot(N2Q_08,av2Q_08,label='double_Q and 0.8')
optimal_value=[0.5 for i in range(NQ_08[-1])]
ax2.plot(optimal_value,label='optimal_value')
ax2.set_xlabel('number of time steps',fontsize=20)
ax2.set_ylabel('average rewards',fontsize=20)
ax2.legend(fontsize=20)

ax3 = fig.add_subplot(223)

ax3.plot(maxQ_1,label='Q and 1')
ax3.plot(max2Q_1,label='double_Q and 1')
optimal_value=[0.36 for i in range(num_play)]
ax3.plot(optimal_value,label='optimal_value')
ax3.set_xlabel('episode',fontsize=20)
ax3.set_ylabel('maximal action_value',fontsize=20)
ax3.legend(fontsize=20)

ax4 = fig.add_subplot(224)

ax4.plot(maxQ_08,label='Q and 0.8')
ax4.plot(max2Q_08,label='double_Q and 0.8')
optimal_value=[0.36 for i in range(num_play)]
ax4.plot(optimal_value,label='optimal_value')
ax4.set_xlabel('episode',fontsize=20)
ax4.set_ylabel('maximal action_value',fontsize=20)
ax4.legend(fontsize=20)

fig.tight_layout()
```
