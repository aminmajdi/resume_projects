Q1. Prove that adding a constant $c$ to all rewards adds a constant ($V_c$)  to the value of all states, and thus does not affect the relative values of any states under any policies. What is $V_c$ in terms of $c$ and $\gamma$. Hint: start from $G_t = \sum_{k=0}^{\infty}\gamma^k R_{t+k+1}$ and calculate $V_{\pi}(s)$.

>Answer:
<br>$V_{\pi}(s) = E_{\pi}(G_t|S_t=s)$ (1)
<br>$G_t = \sum_{k=0}^{\infty}\gamma^k R_{t+k+1}$ (2)
<br>Now if we add constant c to all rewards we have <br>$V^{\prime}_{\pi}(s)$ like this:
<br>$V^{\prime}_{\pi}(s) = E_{\pi}(\sum_{k=0}^{\infty}\gamma^k (R_{t+k+1} + c)|S_t=s)$
<br>$V^{\prime}_{\pi}(s) = E_{\pi}(\sum_{k=0}^{\infty}\gamma^k (R_{t+k+1})|S_t=s)+E_{\pi}(\sum_{k=0}^{\infty}\gamma^k c|S_t=s)$
<br>$V^{\prime}_{\pi}(s) = E_{\pi}(\sum_{k=0}^{\infty}\gamma^k (R_{t+k+1})|S_t=s)+E_{\pi}(\frac{c}{1-\gamma}|S_t=s)$
<br>in the above equation, the first term is exactly like eq. (1).the secound term is made up of c and $\gamma$ and as a result, it remains constant, no matter what policy we choose. Now if we name the secound term as $V_c$ we will have:
<br>$V^{\prime}_{\pi}(s) =V_{\pi}(s)+V_c$ (3)
<br>$V_c=E_{\pi}(\frac{c}{1-\gamma}|S_t=s)=\frac{c}{1-\gamma}$ (4)
<br><br>from (3) and (4) it is obvious that if we add a constant (c) to all the rewards, the relative values will not change:
<br>$V^{\prime}_{\pi}(s_1)-V^{\prime}_{\pi}(s_2)=(V_{\pi}(s_1)+V_c) - (V_{\pi}(s_2)+V_c)=V_{\pi}(s_1)-V_{\pi}(s_2)$
